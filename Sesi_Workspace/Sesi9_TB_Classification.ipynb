````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

````python
# Import dataframe dari CSV file
# Sumber data: # https://www.kaggle.com/rakeshrau/social-network-ads

df = pd.read_csv('Social_Network_Ads.csv')
df = df[['Age', 'EstimatedSalary', 'Purchased']]
df.tail()
````

````python
# Mengecek data apa saja yang ada pada suatu kolom
list(set(list(df['Purchased'])))
#list(set(list(df['Age'])))
````

````python
# Setting: feature variables dan  target variables
feature = ['Age', 'EstimatedSalary']
target = 'Purchased'
len (df)
````

````python
# Splitting training dan test data
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df, test_size=0.2)
len(df_train), len(df_test), len(df)
````

````python
# 1. Preprocessing
# 2. Langsung ke machine learning

from sklearn.preprocessing import StandardScaler

# Scaling data

sc = StandardScaler() # Inisiasi objek StandardScaler

# Buat kolom baru untuk feature variables yang telah diskalakan
# Nanti akan dibandingkan, hasil prediksi menggunakan scaled variables dan unscaled variables
scaled_feature = [x + '_scaled' for x in feature]
scaled_feature
````

````python
sc
````

````python
df_train[feature]
````

````python
sc.fit(df_train[feature])
````

````python
df_train[scaled_feature] = sc.transform(df_train[feature]) # sc.transform eksekusi scaling
df_train
````

````python
df_test[scaled_feature] = sc.transform(df_test[feature])
df_test.head()
````

````python
print(min(df_test['Age']), max(df_test['Age']))
print(min(df_test['Age_scaled']), max(df_test['Age_scaled']))

print(min(df_test['EstimatedSalary']), max(df_test['EstimatedSalary']))
print(min(df_test['EstimatedSalary_scaled']), max(df_test['EstimatedSalary_scaled']))
````

````python
from sklearn.svm import SVC # Import Support Vector Classifier

# Model 1: dengan standard scaller
classifier = SVC(kernel='rbf', random_state=0) # SVC: Support Vector Classifier
classifier.fit(df_train[scaled_feature], df_train[target]) # Target apakah tidak diskalakan?
df_train['Prediction1'] = classifier.predict(df_train[scaled_feature])
df_test['Prediction1'] = classifier.predict(df_test[scaled_feature])
````

````python
# Model 2: tanpa standard scaller
classifier2 = SVC(kernel='rbf', random_state=0)
classifier2.fit(df_train[feature], df_train[target])
df_train['Prediction2'] = classifier2.predict(df_train[feature])
df_test['Prediction2'] = classifier2.predict(df_test[feature])
````

## Classification 
Refs: https://scikit-learn.org/stable/modules/svm.html

SVC (Support Vector Classifier) is class capable of performing binary and multi-class classification on a dataset.

As other classifiers, **SVC, NuSVC, LinearSVC** take as input two arrays:
- an array ```X``` of shape ```(n_samples, n_features)```, holding the training samples, and
- an array ```y``` of class labels (strings or integers), of shape ```(n_samples)```:

````python
from sklearn import svm
X = [[0, 0], [1, 1]] #Input 1, training samples
y = [0, 1] # Input 2, label kelas untuk klasifikasi
clf = svm.SVC() # Inisiasi clf sebagai objek SVC
clf.fit(X, y) # Fitting parameter SVC pada objek clf
````

After being fitted, the model can then be used to predict new values:

````python
clf.predict([[2., 2.]])
````

SVM decision function (detailed in the https://scikit-learn.org/stable/modules/svm.html#svm-mathematical-formulation) depends on some subset of the training data, called the support vectors. Some properties of these support vectors can be found in attributes ```support_vectors_```, ```support_``` and ```n_support_```:

````python
# get support vectors
print(clf.support_vectors_)

# get indices of support vectors
print(clf.support_)

# get number of support vectors for each class
print(clf.n_support_)
````

## Example: SVM: Maximum margin separating hyperplaneÂ¶

refs: https://scikit-learn.org/stable/auto_examples/svm/plot_separating_hyperplane.html#sphx-glr-auto-examples-svm-plot-separating-hyperplane-py

Plot the maximum margin separating hyperplane within a two-class separable dataset using a Support Vector Machine classifier with linear kernel.

````python
print(__doc__)

import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs

# we create 40 separable points
X, y = make_blobs(n_samples=40, centers=2, random_state=6)

# fit the model, don't regularize for illustration purposes
clf = svm.SVC(kernel='linear', C=1000)#rbf
clf.fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

# plot the decision function
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

# plot decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])

# plot support vectors
ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')
plt.show()
````
